plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;

hist — гистограмма, показывающая распределение некоторое величины;

bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.


```python
x = np.linspace(0,10,20)

fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, x ** 2, 'r')
axes.plot(x, x ** 3, 'b*--')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.legend([r'$x^2$', 'x^3'], loc = 0)

plt.show()
```

Нарисуйте график функции x^2, от [-10,10] с шагом 0.5

```python
x = np.arange(-10, 10, 0.5)
plt.plot(x, x**2, label='quadratic')
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.title('x^2')
plt.legend()
plt.grid(True)
plt.show()
```

Теперь график x^2 и x^3 одновременно. На одной картинке.

```python
x = np.arange(-10, 10, 0.5)
plt.ylim(-25, 25)
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.title('x^2 and x^3')
plt.legend()
plt.grid(True)
plt.show()
```

А теперь хочу 2 картинки графиков внутри одной.

```python
x = np.arange(-10, 10, 0.5)
plt.ylim(-25, 25)
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.subplot(211), plt.plot(x, x**2, 'b-')
plt.grid(True)
plt.title('x^2 and x^3')
plt.subplot(212), plt.plot(x, x**3, 'r')
plt.grid(True)
plt.show()
```

```python
x1 = x - x.mean()
y1 = y - y.mean()
b =  sum(x1 * y1)/sum(x1**2)
a = y.mean() - x.mean()*b
a, b
x_plot = np.arange(6, 18)
y_plot = a + b * x_plot
plt.scatter(x, y, color='red')
plt.plot(x_plot, y_plot, '--')
plt.show()
```
